#Identificando y excluyendo outliers

```python
alquiler = pd.read_csv('alquileres.csv', sep = ';')
alquiler.head(5)
```

```python
alquiler.boxplot(['Valor'])
```

```python
alquiler[alquiler['Valor'] >= 500000]
```

```python
valor = alquiler['Valor']
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.25)
Q2 = Q3 - Q1
limite_inferior = Q1 - 1.5 * Q2
limite_superior = Q3 + 1.5 * Q2
```

```python
seleccion = (valor >= limite_inferior) & (valor <= limite_superior)
```

```python
nuevos_datos = alquiler[seleccion]
```

```python
#diagrama de cajas y bigotes
nuevos_datos.boxplot(['Valor'])
```

```python
#histograma de frecuencias
nuevos_datos.hist(['Valor'])
```

```python
grupo_tipo = dataset.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo)
```

```python
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.25)
Q2 = Q3 - Q1
limite_inferior = Q1 - 1.5 * Q2
limite_superior = Q3 + 1.5 * Q2
```

```python
Q1
```

```python
limite_superior
```

```python
datos_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = alquiler['Tipo'] == tipo
  eh_dentro_limite = (alquiler['Valor'] >= limite_inferior[tipo]) & (alquiler['Valor'] <= limite_superior[tipo])
  seleccion = eh_tipo & eh_dentro_limite
  datos_seleccion = alquiler[seleccion]
  datos_new = pd.concat([datos_new, datos_seleccion])
```

```python
datos_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
datos_new.to_csv('alquiler_sin_outliers.csv', sep = ';')
```
